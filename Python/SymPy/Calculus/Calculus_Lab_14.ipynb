```python
import sympy as sp
from sympy.plotting import *
import matplotlib.pyplot as plt
from IPython.display import display,Latex


def contourLine(f, x, y, C):
    return sp.Eq(f(x, y), C)
```

### Задача 2.
Построить на отдельных графиках линии уровня функций $z=\ln(x^2+y)$ и $z=x^2-y^2$.

```python
def func1(x,y):
    return sp.log(x**2 + y)

def func2(x, y):
    return x**2 - y**2


p = sp.plot_implicit(contourLine(func1, x, y, 1), show = False)
for i in [-2, -1, 0, 2, 5]:
    p.append(sp.plot_implicit(contourLine(func2, x, y, i), show = False)[0])
p.show()    
```

### Задача 3.

### Задача 7.

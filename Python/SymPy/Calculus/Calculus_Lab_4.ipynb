## Задание №1
Построить траекторию пули, выпущенной со скоростью $v_0$ под углом $\alpha$ к горизонту. Определить соответствующий диапазон значений параметра времени, учитывая, что пуля не должна провалиться сквозь землю.
$$
\left\{
\begin{matrix}
x=v_0t\cos\alpha\\
y=v_0t\sin\alpha-\frac{gt^2}{2}
\end{matrix}
\right.
$$
Построить несколько траекторий при разных значениях начальной скорости и угла на одном графике. Ускорение свободного падения считать равным 9.8.

```python
v, a, t = symbols("x, a, t")
x = v * t * cos(a)
y = v * t * sin(a) - (9.8 * t * t) /2
x1 = x.subs({v: 70, a: pi / 3})
y1 = y.subs({v: 70, a: pi / 3})
x2 = x.subs({v: 50, a: pi / 6})
y2 = y.subs({v: 50, a: pi / 6})
x3 = x.subs({v: 40, a: pi / 4})
y3 = y.subs({v: 40, a: pi / 4})

plot_parametric((x1, y1), (x2, y2), (x3, y3), (t, 0, 5))
```

## Задание №3
Постройте поверхность в пространстве:
$$
\left\{
\begin{matrix}
x=u\cos u (\cos v +1)\\
y=u\sin u (\cos v +1)\\
z=u\sin v-\left(\frac{u+3}{8}\pi\right)^2-20
\end{matrix}
\right.,
\qquad u\in[0;8\pi],\quad v \in[-\pi;\pi]
$$

```python
from sympy.plotting import  plot3d
u, x, v = symbols("u, x, v")
x = u*cos(u)*(cos(v) + 1)
y = u*sin(u)*(cos(v) + 1)
z = u*sin(v) - ((u + 3) * pi / 8)**2 - 20
plot3d(x, y, z, (u, 0, 8*pi), (v, -pi, pi))
```

## Задание №1
Постройте параболу с вершиной в точке (3;-2), проходящую через начало координат, на интервале [-5;10]. График должен быть зеленой пунктирной линией с маркерами в точках с целочисленными значениями аргумента. Вывести легенду в левом верхнем углу, в легенде написать "Парабола"

```python
from sympy import*
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-15, 15, 512)
y = x**2 - 3*x - 2
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)
plt.grid()
plt.plot(x, y, color = 'green', marker = 'o', linestyle = '--', markerfacecolor = 'green', markevery = 16, label = 'Парабола')
plt.ylim(-10, 10)
plt.xlim(-10, 10)
plt.legend(loc='upper left', ncol=2)
```

## Задание №3
Используя функцию из прошлой лабораторной работы, построить синюю окружность с центром (-1;3) радиуса 2 и отметить на ней маркерами красный треугольник 3 равноудаленные от соседей точки, маркером зеленый квадрат - 4 равноудаленные точки, маркером черный ромб - 6 равноудаленных  точек. Для случая 3 равноудаленных точек подписать верхнюю точку, подпись - координаты в виде символьных выражений. Из этой верхней точки провести пунктирные перпендикуляры черного цвета к осям координат и подписать соответствующие точки на оси их приближенными значениями (по оси х с точностью до десятых, по оси у - сотых). 

```python
def Circle(n, x = -1, y = 3, r = 2):
    x2 = []
    y2 = []
    for i in range(n):
        cor = pi * 2 * i / n
        x2.append(x-r*cos(cor))
        y2.append(y-r*sin(cor))
    return x2, y2
        

plt.grid()
x1, y1 = Circle(350)
plt.ylim(0, 6)
plt.xlim(-5, 4) 
plt.plot(x1, y1, color = 'blue', label = 'Окружность')
x1, y1 = Circle(3)
plt.plot(x1, y1, linestyle = ' ', marker = '^', markerfacecolor = 'red')
coords = " (" + str(x1[2]) + ", " + str(y1[2].evalf(3)) + ")" 
plt.annotate(coords, (x1[2], y1[2]), fontsize = 16) 
plt.annotate(' 4.73', (-5, y1[2]), fontsize = 16) 
plt.annotate(' 0', (x1[2], 0), fontsize = 16) 
plt.plot([-5, 0], [y1[2], y1[2]], 'k--') 
plt.plot([x1[2], x1[2]], [0, y1[2]], 'k--')
x1, y1 = Circle(4)
plt.plot(x1, y1, linestyle = ' ', marker = 's', markerfacecolor = 'green')
x1, y1 = Circle(6)
plt.plot(x1, y1, linestyle = ' ', marker = 'p', markerfacecolor = 'black')
```

### –ó–∞–¥–∞—á–∞ 1
–†–µ—à–∏—Ç—å —É—Ä–∞–≤–Ω–µ–Ω–∏–µ $ùë•^4‚àíe^x=0$, –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫ –ª–µ–≤–æ–π —á–∞—Å—Ç–∏, –æ—Ç–º–µ—Ç–∏—Ç—å —Ç–æ—á–∫—É –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è —Å –æ—Å—å—é –û–•.

```python
x = Symbol('x')
X = np.linspace(-3, 10, 100)
plt.ylim(-20, 20)
plt.plot(X, X**4 - np.exp(X), color = 'red')
plt.plot(X, 0*X, color = 'black')
x1 = nsolve(x**4 - exp(x), x, 0.8)
x2 = nsolve(x**4 - exp(x), x, 0.1)
x3 = nsolve(x**4 - exp(x), x, 8)
plt.grid()
plt.plot([x1, x2, x3], [0, 0, 0], 'vb', color = 'blue')
```

### –ó–∞–¥–∞—á–∞ 2.
–†–µ—à–∏—Ç—å —É—Ä–∞–≤–Ω–µ–Ω–∏–µ $\sin^2(2x)=\ln x$. –ü–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫–∏ –ª–µ–≤–æ–π –∏ –ø—Ä–∞–≤–æ–π —á–∞—Å—Ç–∏, –æ—Ç–º–µ—Ç–∏—Ç—å —Ç–æ—á–∫—É –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–æ–≤.

```python
plt.grid()
Xsin = np.linspace(-4 * np.pi, 4 * np.pi, 100)
Xlog = np.linspace(1, 10, 128)
plt.plot(Xsin, np.sin(Xsin)**2, color = 'green')
plt.plot(Xlog, np.log(Xlog), color = 'red')
x1 = nsolve(sin(x)**2 - log(x), x, 2)
display(x1)
y1 = sin(x1)**2
plt.plot(x1, y1, 'vk', color = 'blue')
```

### –ó–∞–¥–∞—á–∞ 3.
–†–µ—à–∏—Ç—å —É—Ä–∞–≤–Ω–µ–Ω–∏–µ $\sin^2(2x^2-7x+6)=3/4$. –ü–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫–∏ –ª–µ–≤–æ–π –∏ –ø—Ä–∞–≤–æ–π —á–∞—Å—Ç–∏, –æ—Ç–º–µ—Ç–∏—Ç—å —Ç–æ—á–∫—É –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–æ–≤.
–ü—Ä–æ–≤–µ—Ä–∏—Ç—å –ø–æ–¥—Å—Ç–∞–Ω–æ–≤–∫–æ–π.

```python
x, a = symbols('x, a') 
x1 = nsolve(x**2 - S(3) / 4, x, 2) 
x2 = nsolve(x**2 - S(3) / 4, x, -2) 
sinx1 = nsolve(sin(x) - x1, x, 1) 
sinx2 = nsolve(sin(x) - x1, x, 2) 
sinx3 = nsolve(sin(x) - x2, x, -1) 
sinx4 = nsolve(sin(x) - x2, x, -2) 
resx1 = nsolve(2 * x**2 - 7 * x + 6 - sinx1, x, 0.1) 
resx2 = nsolve(2 * x**2 - 7 * x + 6 - sinx2, x, 0.1) 
#resx3 = nsolve(2 * x**2 - 7 * x + 6 - sinx3, x, 000) –Ω–µ—Ç –∫–æ—Ä–Ω–µ–π 
#resx4 = solveset(2*x**2 - 7*x + 6 - sinx4, x, sp.Reals) –Ω–µ—Ç –∫–æ—Ä–Ω–µ–π 
plt.grid() 
x = np.linspace(-1, 4, 1024) 
plt.plot(np.sin(2 * x**2 - 7 * x + 6)**2, color = 'red') 
plt.axhline(y = 0.75, color = 'green', linestyle='-')  
f = sin(2 * a**2 - 7 * a + 6)**2 
print("–†–µ—à–µ–Ω–∏—è:")
display(resx1, resx2)
print("–ü—Ä–æ–≤–µ—Ä–∫–∞:")
display(f.subs({a: resx1})) 
display(f.subs({a: resx2}))
```

### –ó–∞–¥–∞—á–∞ 4.
–†–µ—à–∏—Ç—å —É—Ä–∞–≤–Ω–µ–Ω–∏–µ ${\rm tg}^4(2x^2+1)-1/2=0$, –Ω–∞—á–∞–ª—å–Ω–æ–µ –ø—Ä–∏–±–ª–∏–∂–µ–Ω–∏–µ 0.
–ü—Ä–æ–≤–µ—Ä–∏—Ç—å –ø–æ–¥—Å—Ç–∞–Ω–æ–≤–∫–æ–π.

```python
x, a = symbols('x, a')
x1 = nsolve(x**4 - S(1) / 2, x, 1) 
x2 = nsolve(x**4 - S(1) / 2, x, -1) 
tang = nsolve(tan(x) - x1, x, 2) 
res = nsolve(2 * x**2 + 1 - tang, x, 1.5) 
f = tan(2 * a**2 + 1)**4 
print("–†–µ—à–µ–Ω–∏–µ:")
display(tang) 
print("–ü—Ä–æ–≤–µ—Ä–∫–∞:")
display(f.subs({a: res}))
```

### –ó–∞–¥–∞—á–∞ 5.
–†–µ—à–∏—Ç—å —Å–∏—Å—Ç–µ–º—É —É—Ä–∞–≤–Ω–µ–Ω–∏–π 
$$
\left\{\begin{matrix}
(x-1)^2+(y-2)^2+(z+3)^2=25\\
x+y+z=0\\
-2x+3y-z-7=0
\end{matrix}\right.
$$
–Ω–∞—á–∞–ª—å–Ω–æ–µ –ø—Ä–∏–±–ª–∏–∂–µ–Ω–∏–µ (0,0,0).
–ü—Ä–æ–≤–µ—Ä–∏—Ç—å –ø–æ–¥—Å—Ç–∞–Ω–æ–≤–∫–æ–π.

```python
x, y, z = sp.symbols('x y z') 
f1 = (x - 1)**2 + (y - 2)**2 + (z + 3)**2 - 25 
f11 = (x - 1)**2 + (y - 2)**2 + (z + 3)**2 
f2 = x + y + z 
f3 = -2*x + 3*y - z - 7 
f31 = -2*x + 3*y - z 
res = sp.nsolve((f1, f2, f3), (x, y, z), (0, 0, 0)) 
display(res) 
print("–ü–æ–¥—Å—Ç–∞–Ω–æ–≤–∫–∞:")
display(f11.subs({x: res[0], y: res[1], z: res[2]}), f31.subs({x: res[0], y: res[1], z: res[2]})) 
```

### –ò–Ω–¥–∏–≤–∏–¥—É–∞–ª—å–Ω–æ–µ –∑–∞–¥–∞–Ω–∏–µ
–ü–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫–∏ —Ñ—É–Ω–∫—Ü–∏–π, —É—Ç–æ—á–Ω–∏—Ç—å –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã —Ç–æ—á–µ–∫ –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è, —Ä–µ—à–∞—è —á–∏—Å–ª–µ–Ω–Ω–æ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–µ–µ —É—Ä–∞–≤–Ω–µ–Ω–∏–µ –∏–ª–∏ —Å–∏—Å—Ç–µ–º—É —É—Ä–∞–≤–Ω–µ–Ω–∏–π. –ù–∞  –≥—Ä–∞—Ñ–∏–∫–µ –æ—Ç–º–µ—Ç–∏—Ç—å –∏ –ø–æ–¥–ø–∏—Å–∞—Ç—å –±—É–∫–≤–∞–º–∏ –ê,–í,... —Ç–æ—á–∫–∏ –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–æ–≤.  

```python
#f(x) = 2x2 + 3x + 6, g(x) = cos(4x2 + 44x + 3) + 10
x, x1 = symbols('x x_1') 
g = cos(4 * x1**2 + 44 * x1 + 3) + 10 
f = 2 * x1**2 + 3 * x1 + 6 
xArr = [0, 0, 0, 0, 0, 0, 0, 0, 0] 
xArr[0] = nsolve(g - f, x1, -2.5) 
xArr[1] = nsolve(g - f, x1, -2.45) 
xArr[2] = nsolve(g - f, x1, -2) 
xArr[3] = nsolve(g - f, x1, 0.78) 
xArr[4] = nsolve(g - f, x1, 0.813) 
xArr[5] = nsolve(g - f, x1, 0.85) 
xArr[6] = nsolve(g - f, x1, 0.95) 
xArr[7] = nsolve(g - f, x1, 0.995) 
x = np.linspace(-3.0, 1.5, 1028)
plt.grid()
plt.ylim(8.9, 11.1)
plt.plot(x, np.cos(4 * x**2 + 44 * x + 3) + 10) 
plt.plot(x, 2 * x**2 + 3 * x + 6)

plt.annotate(r'$(H)$',
             xy = (xArr[7], 2 * xArr[7]**2 + 3 * xArr[7] + 6 ), xycoords = 'data', 
             fontsize = 16, xytext = (-10, -50), textcoords = 'offset points',
             arrowprops = dict(arrowstyle = "->", connectionstyle = "arc3,rad=.2"))

plt.annotate(r'$(G)$',
             xy = (xArr[6], 2 * xArr[6]**2 + 3 * xArr[6] + 6 ), xycoords = 'data', 
             fontsize = 16, xytext = (-10, -50), textcoords = 'offset points',
             arrowprops = dict(arrowstyle = "->", connectionstyle = "arc3,rad=.2"))

plt.annotate(r'$(F)$',
             xy = (xArr[5], 2 * xArr[5]**2 + 3 * xArr[5] + 6 ), xycoords = 'data', 
             fontsize = 16, xytext = (-10, -50), textcoords = 'offset points',
             arrowprops = dict(arrowstyle = "->", connectionstyle = "arc3,rad=.2"))

plt.annotate(r'$(E)$',
             xy = (xArr[4], 2 * xArr[4]**2 + 3 * xArr[4] + 6 ), xycoords = 'data', 
             fontsize = 16, xytext = (-10, -50), textcoords = 'offset points',
             arrowprops = dict(arrowstyle = "->", connectionstyle = "arc3,rad=.2"))

plt.annotate(r'$(D)$',
             xy = (xArr[3], 2 * xArr[3]**2 + 3 * xArr[3] + 6 ), xycoords = 'data', 
             fontsize = 16, xytext = (-10, -50), textcoords = 'offset points',
             arrowprops = dict(arrowstyle = "->", connectionstyle = "arc3,rad=.2"))

plt.annotate(r'$(C)$',
             xy = (xArr[2], 2 * xArr[2]**2 + 3 * xArr[2] + 6 ), xycoords = 'data', 
             fontsize = 16, xytext = (-10, -50), textcoords = 'offset points',
             arrowprops = dict(arrowstyle = "->", connectionstyle = "arc3,rad=.2"))

plt.annotate(r'$(B)$',
             xy = (xArr[1], 2 * xArr[1]**2 + 3 * xArr[1] + 6 ), xycoords = 'data', 
             fontsize = 16, xytext = (-10, -50), textcoords = 'offset points',
             arrowprops = dict(arrowstyle = "->", connectionstyle = "arc3,rad=.2"))

plt.annotate(r'$(A)$',
             xy = (xArr[0], 2 * xArr[0]**2 + 3 * xArr[0] + 6 ), xycoords = 'data', 
             fontsize = 16, xytext = (-10, -50), textcoords = 'offset points',
             arrowprops = dict(arrowstyle = "->", connectionstyle = "arc3,rad=.2"))
```

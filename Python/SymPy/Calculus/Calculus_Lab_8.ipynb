## Задание №1
Исследовать функцию с помощью производной, построить график функции, отметить и подписать точки экстремума и точки перегиба. 
$$
\left\{
\begin{matrix}
x=3cos(t)\\
y=5sin(t)
\end{matrix}
\right.
$$

```python
from sympy import *
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
init_printing(use_unicode=False, wrap_line=False, no_global=True)
```

```python
t = Symbol('t')
X = cos(t) * 3
Y = sin(t) * 5
dyx = diff(Y, t) / diff(X, t)
dy2xx = (diff(X, t) * diff(Y, t,2) - diff(X, t,2) * diff(Y, t)) / (diff(X, t))**3
(dyx, dy2xx)
dyxNum, dyxDen = fraction(dyx)
dyxNumRes = solveset(dyxNum)
dyxDenRes = solveset(dyxDen)
dyxPoints = list((dyxNumRes.union(dyxDenRes)).intersect(Interval(0 , 2 * pi)))
dyxPoints.sort()
#display(dy_x_points)
points=[]
for i in range(len(dyxPoints) - 1):
    points += [(dyxPoints[i] + dyxPoints[i + 1]) / 2]
#display(points)  
maxArr=[]
minArr=[]
for i in range(1, len(dyxPoints) - 1):
    if dyx.subs(t, points[i - 1]) > 0 and dyx.subs(t, points[i]) < 0:
        if X.subs(t, points[i - 1]) < X.subs(t, points[i]):
            if dyxPoints[i] not in dyxDenRes:
                maxArr+=[dyxPoints[i]]
        else:
            if dyxPoints[i] not in dyxDenRes:
                minArr += [dyxPoints[i]]
    elif  dyx.subs(t, points[i - 1]) < 0 and dyx.subs(t, points[i]) > 0:  
        if X.subs(t, points[i - 1]) > X.subs(t, points[i]):
            if dyxPoints[i] not in dyxDenRes:
                maxArr += [dyxPoints[i]]
        else:
            if dyxPoints[i] not in dyxDenRes:
                minArr+=[dyxPoints[i]]
print("Значения t в точке максимума на интервале [0, 2pi]")
display(maxArr)
print("Значения t в точке максимума на интервале [0, 2pi]")
display(minArr)
print("график функции")
T = np.linspace(0, 2 * np.pi, 1024)
plt.plot(np.cos(T) * 3, np.sin(T) * 5)
plt.axis('equal')
ax = plt.gca()  # gca stands for 'get current axis'
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))
for item in maxArr:
    plt.scatter([X.subs(t,item)],[Y.subs(t,item)], 50, color='red')
    plt.annotate(r'$\max:\ t='+latex(item)+',\,x='+latex(X.subs(t,item))+',\,y='+latex(Y.subs(t,item))+r'$',xy=(X.subs(t,item),Y.subs(t,item)), xycoords='data',
             xytext=(40,30), textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))
for item in minArr:
    plt.scatter([X.subs(t,item)],[Y.subs(t,item)], 50, color='red')
    plt.annotate(r'$\min:\ t='+latex(item)+',\,x='+latex(X.subs(t,item))+',\,y='+latex(Y.subs(t,item))+r'$',xy=(X.subs(t,item),Y.subs(t,item)), xycoords='data',
             xytext=(20,-30), textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))
```

## Задание №2
Исследовать функцию с помощью производной, построить график функции, отметить и подписать точки экстремума и точки перегиба. 
$$
\left\{
\begin{matrix}
x=sin(2t)\\
y=sin(3t)
\end{matrix}
\right.
$$

```python
t = Symbol('t')
X = sin(2 * t)
Y = sin(3 * t)
dyx = diff(Y, t) / diff(X, t)
dy2xx = (diff(X, t) * diff(Y, t, 2) - diff(X, t, 2) * diff(Y, t)) / (diff(X, t))**3

dy2xxNum, dy2xxDen = fraction(dy2xx)
dy2xxNumRes = solveset(dy2xxNum)
dy2xxDenRes = solveset(dy2xxDen)
dy2xxPoints = list((dy2xxNumRes.union(dy2xxDenRes)).intersect(Interval(0, 2 * pi + 0.1)))
dy2xxPoints.sort()
display(dy2xxPoints)
points = []
for i in range(len(dy2xxPoints) - 1):
    points += [(dy2xxPoints[i] + dy2xxPoints[i + 1]) / 2]
display(points)  
infArr = []
for i in range(1, len(dy2xxPoints) - 1):
    if dy2xx.subs(t, points[i - 1]) * dy2xx.subs(t, points[i]) < 0:
        infArr += [dy2xxPoints[i]]
display(infArr)
plt.plot(np.sin(2*T),np.sin(3*T))
plt.axis('equal')
xcount = -50
ycount = 20
for item in infArr:
    plt.scatter([X.subs(t,item)],[Y.subs(t,item)], 50, color='red')
    plt.annotate(r'$\left('+latex(item)+','+latex(X.subs(t,item))+','+latex(Y.subs(t,item))+r'\right)$',xy=(X.subs(t,item),Y.subs(t,item)), xycoords='data',
             xytext=(50+xcount,20+ycount), textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))
    xcount+=10
    ycount+=10
```

### Задача 1
Вычислить площадь фигуры, ограниченной графиками функций $y=2\sin \sqrt[]{x} \quad y=e^{1/x}$ и правой ветвью параболы $y=4-\frac{x^{2}}{2}$ 

```python
x, f, t = sp.symbols("x, φ, t")

y1 = 2 * sp.sin(sp.sqrt(x))
y2 = sp.E**(1 / x)
y3 = 4 - x**2 / 2

display(y1, y2, y3)
```

```python
def plot_mult(sym, range1, range2, *func):
    d = []
    for i in func:
        d.append((i, (sym, range1, range2)))
    sp.plot(*d)
    
plot_mult(x, 0.4, 2, y1, y2, y3)
```

```python
fn = sp.Max(y1, y2) - y3
sp.plot(fn, (x, 0.6, 2.1))
```

```python
p1 = sp.nsolve(y1-y3, x, 2)
p2 = sp.nsolve(y2-y3, x, 0.1)
p3 = sp.nsolve(y1 - y2, x, 1.4) # середина
display(p1, p2, p3)
```

```python
s1 = abs(float(sp.integrate(y1 - y3, (x, p1, p2))))
s2 = abs(float(sp.integrate(y2 - y3, (x, p2, p3))))
s1 + s2
```

### Задача 2.
Вычислить площадь фигуры, ограниченной графиком функции

$ x=\cos ^{3} \left( t \right) , y=\sin ^{3} \left( t \right) , 0 \leq t \leq 2 \pi $

```python
x = sp.cos(t) ** 3
y = sp.sin(t) ** 3
```

```python
plot_mult(t, -4, 4, x, y)
```

```python
p1 = sp.nsolve(x - y, t,  -2.3)
p2 = sp.nsolve(x - y, t, 0.9)
p1, p2
```

```python
sp.integrate(x-y, (t, p1, p2))
```

### Задача 3.

$ r=cos \left(  \varphi  \right) , -\frac{ \pi }{2} \leq  \varphi  \leq \frac{ \pi }{2} $

```python
r = sp.cos(f)

s = 1 / 2 * sp.integrate(r**2, (f, -sp.pi / 2, sp.pi / 2))
s
```

### Задача 4.
Вычислить длину кривой, заданной уравнением

$ y=x^{3/2} $

```python
x = sp.symbols("x")
y = x**(3 / 2)
sp.plot(y)

sp.integrate(sp.sqrt(1 + (y.diff(x))**2), (x, 0, 10))
```

### Задача 5.
Вычислить длину кривой, заданной параметрически

$ x=e^{t}\cos t, y=e^{t}\sin t, 0 \leq t \leq 1 $

```python
x = sp.E**t * sp.cos(t)
y = sp.E**t * sp.sin(t)

curve = sp.Curve((x, y), (t, 0, 1))
curve.length
```

### Задача 6.
Вычислить длину кривой, заданной в полярных координатах

### Задание 1.

Составить в матричном виде двумя способами уравнение поверхности второго порядка $4x^2-24xy+8\sqrt3 xz-7y^2+10\sqrt3 yz+3z^2-4$. Привести уравнение поверхности второго порядка к каноническому виду с помощью перехода к новому базису.

Уравнение поверхности второго порядка:
$$
a_{11}x^2+a_{22}y^2+a_{33}z^2+2a_{12}xy+2a_{13}xz+2a_{23}yz+2a_{14}x+2a_{24}y+2a_{34}z+a_{44}
$$
В матричном виде:
$$
X^TAX+2bX+c,\quad X=\left(\begin{matrix}x\\y\\z\end{matrix}\right),\quad  A=\left(\begin{matrix}a_{11}&a_{12}&a_{13}\\a_{12}&a_{22}&a_{23}\\a_{13}&a_{23}&a_{33}\end{matrix}\right),\quad b=\left(\begin{matrix}a_{14}&a_{24}&a_{34}\end{matrix}\right),\quad c=a_{44}
$$

$$
X^TA_1X,\quad X=\left(\begin{matrix}x\\y\\z\\1\end{matrix}\right), A=\left(\begin{matrix}a_{11}&a_{12}&a_{13}&a_{14}\\a_{12}&a_{22}&a_{23}&a_{24}\\a_{13}&a_{23}&a_{33}&a_{34}\\a_{14}&a_{23}&a_{33}&a_{44}\end{matrix}\right)
$$


```python
x, y, z = sp.symbols('x y z')


A = sp.Matrix([[4, -12, 4 * sp.sqrt(3)], [-12, -7, 5 * sp.sqrt(3)], [4 * sp.sqrt(3), 5 * sp.sqrt(3), 3]])
b = sp.Matrix((0, 0, 0)).T
c = -4
X1 = sp.Matrix((x, y, z))

display(A, sp.simplify(sp.expand(X1.T * A * X1))[0])
res = A.eigenvects()
P = res[0][-1][0].row_join(res[1][-1][0]).row_join(res[2][-1][0])
A1 = P.T * A * P
display(A1, sp.simplify(sp.expand(X1.T * A1 * X1))[0] - 4)
```

### Задание 2.
Привести уравнение поверхности второго порядка $- 4 x^{2} - 16 \sqrt{2} x y - 16 \sqrt{2} x z + 2 y^{2} - 12 y z + 2 z^{2} - 4$ к каноническому виду. Построить график.

```python
A = sp.Matrix([[-4, -8 * sp.sqrt(2), -8 * sp.sqrt(2)], [-8 * sp.sqrt(2), 2, -6], [-8 * sp.sqrt(2), -6, 2]])
b = sp.Matrix((0, 0, 0)).T
c = -4
X1 = sp.Matrix((x, y, z))
display(A, sp.simplify(sp.expand(X1.T * A * X1))[0])
res = A.eigenvects()
P = res[0][-1][0].row_join(res[1][-1][0]).row_join(res[2][-1][0])
A1 = P.T * A * P
display(A1, sp.simplify(sp.expand(X1.T * A1 * X1))[0]-4)
plot3d(-sp.sqrt(-80 * x**2 + 16 * y**2 - 4) / sp.sqrt(48), sp.sqrt(-80 * x**2 + 16 * y**2 - 4) / sp.sqrt(48))
```
